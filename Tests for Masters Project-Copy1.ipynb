```python
#Converting all rorws into a list of tuples
y = []
for i in range(len(df)):
    x = tuple(df.iloc[i])
    y.append(x)
    
y
```

```python
#Save the list as txt file
file = open('Evapotranspiracao.txt', 'w')

for tuple in y:
    file.write(str(tuple)+','+'\n')

file.close()
```
